```python
def uniplot(inp0,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 21
    plt.rcParams['axes.titlesize'] = 23
    plt.rcParams['axes.titlepad'] = 30
    
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(inp0[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = inp0, x= col, order=inp0[col].value_counts().index,hue = hue,palette='viridis')
    
    plt.show()
```

```python
# Plotting the above correlation with heat map as it is the best choice to visulaize

# figure size

def targets_corr(data,title):
    plt.figure(figsize=(15, 10))
    plt.rcParams['axes.titlesize'] = 25
    plt.rcParams['axes.titlepad'] = 70

# Heatmap with a color map of choice


    sns.heatmap(data, cmap="RdYlGn",annot=False)

    plt.title(title)
    plt.yticks(rotation=0)
    plt.show()
```
